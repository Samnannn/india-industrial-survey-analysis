```python
dfs = {
    "dfa": dfa,
    "dfb": dfb,
    "dfc": dfc,
    "dfd": dfd,
    "dfe": dfe,
    "dff": dff,
    "dfg" : dfg,
    "dfh" : dfh,
    "dfi" : dfi,
    "dfj" : dfj,
    "dfk" : dfk,
    "dfl" : dfl,
    "dfm" : dfm,
    "dfn" : dfn
}

base_cols = set(dfa.columns)

for name, df in dfs.items():
    if set(df.columns) != base_cols:
        print(f"❌ {name} has different columns")
    else:
        print(f"✅ {name} columns match")
```

```python
for name, df in dfs.items():
    missing = base_cols - set(df.columns)
    extra = set(df.columns) - base_cols

    if missing or extra:
        print(f"\n{name}")
        print("Missing columns:", missing)
        print("Extra columns:", extra)
    else:
        print(f"{name} ✅ all good")
```

```python
nic = pd.read_csv("./data/2011/nic_2008.csv")
```

```python
final_df = final_df.merge(nic, on="Ind_5digit", how="left")
```

```python
cols = list(final_df.columns)

# remove desc from current position
cols.remove("desc")

# find position of Ind_5digit
ind_pos = cols.index("Ind_5digit")

# insert desc right after it
cols.insert(ind_pos + 1, "desc")

# reorder dataframe
final_df = final_df[cols]
```

```python
state = pd.read_csv("./data/2011/tabula-state1.csv")
```

```python
final_df = final_df.merge(state, on="State", how="left")
```

```python
cols = list(final_df.columns)

# remove desc from current position
cols.remove("State_code")

# find position of Ind_5digit
state_pos = cols.index("State")

# insert desc right after it
cols.insert(state_pos + 1, "State_code")

# reorder dataframe
final_df = final_df[cols]
```

```python
final_df["Sector_code"] = final_df["Sector"].map({
    "Rular": 1,
    "Urban": 2
})
```

```python
cols = list(final_df.columns)

# remove desc from current position
cols.remove("Sector_code")

# find position of Ind_5digit
stector_pos = cols.index("Sector")

# insert desc right after it
cols.insert(stector_pos + 1, "Sector_code")

# reorder dataframe
final_df = final_df[cols]
```

```python
final_df['Status'].unique()
```

```python
final_df['Status'] = (
    final_df['Status']
        .str.strip()                      # remove leading & trailing spaces
        .str.replace(r'\s+', ' ', regex=True)   # remove extra spaces inside
        .str.replace('?', '-', regex=False)     # fix weird ? symbol
)
```

```python
final_df['Status'] = (
    final_df['Status']
        .str.strip()                      # remove start & end spaces
        .str.replace(r'\s+', ' ', regex=True)   # replace multiple spaces with single space
)
```

```python
final_df['Status'].unique()
```

```python
status = pd.read_excel("./data/2011/status (2).xlsx")
```

```python
final_df = final_df.merge(status, on="Status", how="left")
```

```python
cols = list(final_df.columns)

# remove desc from current position
cols.remove("Status_code")

# find position of Ind_5digit
status_pos = cols.index("Status")

# insert desc right after it
cols.insert(status_pos + 1, "Status_code")

# reorder dataframe
final_df = final_df[cols]
```

```python
final_df['State'] = final_df['State'].replace('Telangana', 'Andhra Pradesh')
```

```python
final_df.loc[final_df['State_code'] == 36, 'State_code'] = 28
```

```python
column_rename_mapping = {
    'Non_manu_man_day_male': 'Non_manu_man_day_Male',
    'Manu_man_day_male': 'Manu_man_day_Male',
    'Total_man_day_male': 'Total_man_day_Male'
}

# Rename the columns if they exist in the DataFrame
final_df = final_df.rename(columns=column_rename_mapping)
```
