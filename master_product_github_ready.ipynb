```python
dfs = {
    "dfa": dfa,
    "dfb": dfb,
    "dfc": dfc,
    "dfd": dfd,
    "dfe": dfe,
    "dff": dff,
    "dfg" : dfg,
    "dfh" : dfh,
    "dfi" : dfi,
    "dfj" : dfj,
    "dfk" : dfk,
    "dfl" : dfl,
    "dfm" : dfm,
    "dfn" : dfn
}

base_cols = set(dfa.columns)

for name, df in dfs.items():
    if set(df.columns) != base_cols:
        print(f"❌ {name} has different columns")
    else:
        print(f"✅ {name} columns match")
```

```python
for name, df in dfs.items():
    missing = base_cols - set(df.columns)
    extra = set(df.columns) - base_cols

    if missing or extra:
        print(f"\n{name}")
        print("Missing columns:", missing)
        print("Extra columns:", extra)
    else:
        print(f"{name} ✅ all good")
```

```python
# clean all description columns
final_df['desc'] = final_df['desc'].str.strip().str.lower()

asicc_mis1['desc_old'] = asicc_mis1['desc_old'].str.strip().str.lower()
asicc_mis1['desc_rev'] = asicc_mis1['desc_rev'].str.strip().str.lower()

# create mapping using BOTH old and new desc
old_map = dict(zip(asicc_mis1['desc_old'], asicc_mis1['asicc']))
new_map = dict(zip(asicc_mis1['desc_rev'], asicc_mis1['asicc']))

# combine both mappings
combined_map = {**old_map, **new_map}

# fill only NaN ASICC values
final_df.loc[final_df['asicc'].isna(), 'asicc'] = (
    final_df.loc[final_df['asicc'].isna(), 'desc']
    .map(combined_map)
)
```
