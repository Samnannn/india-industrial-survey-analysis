```python
def compare_columns(df1, df2, name1="df1", name2="df2"):
    cols1 = set(df1.columns)
    cols2 = set(df2.columns)

    only_in_df1 = sorted(cols1 - cols2)
    only_in_df2 = sorted(cols2 - cols1)

    print(f"\n{name1} columns: {len(cols1)}")
    print(f"{name2} columns: {len(cols2)}")

    print(f"\nOnly in {name1} ({len(only_in_df1)}):")
    print(only_in_df1)

    print(f"\nOnly in {name2} ({len(only_in_df2)}):")
    print(only_in_df2)

    print(f"\nCommon columns: {len(cols1 & cols2)}")

    return only_in_df1, only_in_df2
```
